```python
import pandas as pd
import requests
import re
import cleaning_functions as cf
```

```python
df_golden = pd.read_csv('../data/raw/golden_globe_awards.csv')
```

```python
df_golden.drop('ceremony',axis=1,inplace=True)
df_golden.head(2)
```

```python
df_golden.category = df_golden.category.str.capitalize()
```

```python
#eliminamos los mismos años que en oscars
df_golden = df_golden[df_golden.year_film >= 1967]
df_golden
```

```python
#Parece que hay categorías que no tienen una película vinculada, así que vamos a hacer una primera limpieza de los valores en Category para quitarnos las innecesarias
```

```python
df_golden.category.unique()
```

```python
#primero eliminamos las filas con categorías sobre televisión, ya que nuestro estudio se centra en cine/películas
tv_categories = cf.values_iterable_regex(df_golden.category,r'\b\w*television\w*\b')
```

```python
df_golden = cf.drop_rows_by_value(df_golden,df_golden.category,tv_categories)
```

```python
#eliminamos también categrías honoríficas no vinculadas a una película concreta, y las categrías "New star" que se eliminaron en los 80:
honorary_awards =['Cecil b. demille award','Henrietta award (world film favorites)','Special achievement award','Carol burnett award','New star of the year - actress', 'New star of the year - actor', 'New star of the year']
```

```python
df_golden = cf.drop_rows_by_value(df_golden,df_golden.category,honorary_awards)
```

```python
cat_changes = {
    'Actress in a leading role - musical or comedy': 'Best actress - musical or comedy',
    'Best performance by an actress in a motion picture - drama': 'Best actress - drama',
    'Best performance by an actress in a motion picture - musical or comedy': 'Best actress - musical or comedy',
    'Best performance by an actress in a supporting role in any motion picture': 'Best supporting actress - motion picture',
    'Best performance by an actor in a motion picture - drama': 'Best actor - drama',
    'Best performance by an actor in a motion picture - musical or comedy': 'Best actor - musical or comedy',
    'Best performance by an actor in a supporting role in any motion picture': 'Best supporting actor - motion picture',
    'Foreign film - foreign language': 'Best motion picture - foreign language',
    'Foreign film - english language': 'Best motion picture - foreign language'
}
```

```python
df_golden = cf.clean_values_column (df_golden, cat_changes)
df_golden.category.unique()
```
