```python
import pandas as pd
import re
import cleaning_functions as cf
```

```python
golden_scrap =pd.read_csv('../data/Clean/golden_scrap_cleaned.csv')
```

```python
golden_df = pd.read_csv('../data/Clean/golden_cleaned.csv')
```

```python
df_golden_final = pd.concat([golden_df,golden_scrap], ignore_index = True)
```

```python
#los títulos que empiezan por 'The' lo tienen al final después de una coma. Este formato es distinto que el del df de los oscars e IMDb
```

```python
updated_film = []
film_list = df_golden_final.film.to_list()
for x in df_golden_final.film:
    if re.search(r'\b\w*, The', x, flags=re.IGNORECASE):
        y = 'The '+x
        updated_film.append(y)
    else:
        updated_film.append(x)

values_replace_dictio = dict(zip(film_list,updated_film))
df_golden_final = cf.clean_values_column (df_golden_final, values_replace_dictio)
df_golden_final.film = df_golden_final.film.str.replace(', The','')
```

```python
count = 0
for x in df_golden_final.film:
    if re.search(r'\b\w*, The', x, flags=re.IGNORECASE):
        count += 1
count
```
