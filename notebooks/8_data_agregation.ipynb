```python
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
nominated_movies = pd.read_csv('../data/Clean/4_final_movie_data.csv')
nominated_movies.head(2)
```

```python
#comparación de películas nominadas en ambos galardones vs el total
```

```python
nominations_years = nominated_movies.groupby('year_film')['film'].agg('count')
```

```python
nominated_both = nominated_movies[(nominated_movies['Total oscars nominations'] > 0) & (nominated_movies['G.G. nominations']>0)]
nominated_both = nominated_both.groupby('year_film')['film'].agg('count')
```

```python
comp_nominations_years = pd.merge(nominations_years,nominated_both, how='outer', on = ['year_film'])
comp_nominations_years.fillna(0, inplace = True)
```

```python
comp_nominations_years = comp_nominations_years.rename(columns = {'film_x':'Total nominations','film_y':'Nominated both awards'})
```

```python
comp_nominations_years[['Total nominations','Nominated both awards']].plot()
plt.xlabel('Years')
plt.ylabel('N. films')
plt.title('Nominated films all categories')
```

```python
comp_nominations_years.iloc[1,0]
```

```python
#media de coincidencia
average = []
for y in range(len(comp_nominations_years)):
    perc = int(comp_nominations_years.iloc[y,1])/int(comp_nominations_years.iloc[y,0])
    average.append(perc)
sum(average)/len(average)*100
    
    
```

```python
#estudiamos la valoración media de las películas, sacamos las películas nominadas que tienen valoración y repetimos la selección anterior
#en este caso hacemos la comparación de la mediana de la valoración para cada grupo, separando Oscars y Globos de Oro para ver cual es más fiable
```

```python
nominated_movies_score = nominated_movies[pd.to_numeric(nominated_movies['TMDb Score'], errors='coerce').notnull()]
```

```python
nominated_movies_score['TMDb Score'] = nominated_movies_score['TMDb Score'].astype(float)
```

```python
values = nominated_movies_score.describe()
nominated_median_value = values.loc['50%','TMDb Score']
```

```python
#películas nominadas, todas las categorías:
```

```python
nominated_oscars_score = nominated_movies_score[(nominated_movies_score['Total oscars nominations'] > 0)]
nominated_oscars_score = nominated_oscars_score.groupby('year_film')['TMDb Score'].agg('median')
```

```python
nominated_golden_score = nominated_movies_score[(nominated_movies_score['G.G. nominations'] > 0)]
nominated_golden_score = nominated_golden_score.groupby('year_film')['TMDb Score'].agg('median')
```

```python
nominations_scores = pd.merge(nominated_oscars_score,nominated_golden_score, how='outer', on = ['year_film'])
nominations_scores = nominations_scores.rename(columns = {'TMDb Score_x':'Score Oscars','TMDb Score_y':'Score Golden Globes'})
nominated_median_score_all = []
for l in range(len(nominations_scores)):
    nominated_median_score_all.append(nominated_median_value)
nominations_scores['All films median score'] = nominated_median_score_all
```

```python
nominations_scores[['Score Oscars','Score Golden Globes','All films median score']].plot()
plt.xlabel('Years')
plt.ylabel('Median score')
plt.title('Nominated films scores')
```

```python
#películas ganadoras, todas las categorías:
```

```python
winners_oscars_score = nominated_movies_score[(nominated_movies_score['Total oscars won'] > 0)]
winners_oscars_score = winners_oscars_score.groupby('year_film')['TMDb Score'].agg('median')
winners_golden_score = nominated_movies_score[(nominated_movies_score['G.G. won'] > 0)]
winners_golden_score = winners_golden_score.groupby('year_film')['TMDb Score'].agg('median')
```

```python
winners_scores = pd.merge(winners_oscars_score,winners_golden_score, how='outer', on = ['year_film'])
winners_scores.head(1)
```

```python
winners_scores = winners_scores.rename(columns = {'TMDb Score_x':'Score Oscars winners','TMDb Score_y':'Score Golden Globes winners'})
nominated_median_score_all = []
winners_median_score_all = []
winners_median = nominated_movies_score[(nominated_movies_score['Total oscars won'] > 0) | (nominated_movies_score['G.G. won']>0)]['TMDb Score'].mean()
for l in range(len(winners_scores)):
    nominated_median_score_all.append(nominated_median_value)
    winners_median_score_all.append(winners_median)
winners_scores['All films median score'] = nominated_median_score_all
winners_scores['All winners median score'] = winners_median_score_all
```

```python
winners_scores[['Score Oscars winners','Score Golden Globes winners','All films median score','All winners median score']].plot()
plt.xlabel('Years')
plt.ylabel('Median score')
plt.title('Awarded films scores')
```

```python
#películas nominadas, categorías principales:
```

```python
categories_oscars_score = nominated_movies_score[(nominated_movies_score['Oscars main categories nominations'] > 0)]
categories_oscars_score = categories_oscars_score.groupby('year_film')['TMDb Score'].agg('median')
categories_golden_score = nominated_movies_score[(nominated_movies_score['G.G. main categories nominations'] > 0)]
categories_golden_score = categories_golden_score.groupby('year_film')['TMDb Score'].agg('median')
```

```python
categories_scores = pd.merge(categories_oscars_score,categories_golden_score, how='outer', on = ['year_film'])
categories_scores = categories_scores.rename(columns = {'TMDb Score_x':'Score Oscars main cat.','TMDb Score_y':'Score Golden Globes main cat.'})
categories_scores.head(1)
```

```python
nominated_median_score_all = []
categories_median_score_all = []
categories_median = nominated_movies_score[(nominated_movies_score['Oscars main categories nominations'] > 0) | (nominated_movies_score['G.G. main categories nominations']>0)]['TMDb Score'].mean()
for l in range(len(categories_scores)):
    nominated_median_score_all.append(nominated_median_value)
    categories_median_score_all.append(categories_median)
categories_scores['All films median score'] = nominated_median_score_all
categories_scores['Main cat. nominated median score'] = categories_median_score_all
```

```python
categories_scores[['Score Oscars main cat.','Score Golden Globes main cat.','All films median score','Main cat. nominated median score']].plot()
plt.xlabel('Years')
plt.ylabel('Median score')
plt.title('Main categories nominated films scores')
```
