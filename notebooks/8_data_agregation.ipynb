```python
comp_nominations_years = comp_nominations_years.rename(columns = {'film_x':'Total nominations','film_y':'Nominated both awards'})
```

```python
#estudiamos la valoración media de las películas, sacamos las películas nominadas que tienen valoración y repetimos la selección anterior
#en este caso hacemos la comparación de la mediana de la valoración para cada grupo, separando Oscars y Globos de Oro para ver cual es más fiable
```

```python
nominated_movies_score = nominated_movies[pd.to_numeric(nominated_movies['TMDb Score'], errors='coerce').notnull()]
```

```python
nominated_movies_score['TMDb Score'] = nominated_movies_score['TMDb Score'].astype(float)
```

```python
values = nominated_movies_score.describe()
nominated_median_value = values.loc['50%','TMDb Score']
```

```python
#películas nominadas, todas las categorías:
```

```python
nominated_oscars_score = nominated_movies_score[(nominated_movies['Total oscars nominations'] > 0)]
nominated_oscars_score = nominated_oscars_score.groupby('year_film')['TMDb Score'].agg('median')
```

```python
nominated_golden_score = nominated_movies_score[(nominated_movies['G.G. nominations'] > 0)]
nominated_golden_score = nominated_golden_score.groupby('year_film')['TMDb Score'].agg('median')
```

```python
nominations_scores = pd.merge(nominated_oscars_score,nominated_golden_score, how='outer', on = ['year_film'])
nominations_scores = nominations_scores.rename(columns = {'TMDb Score_x':'Score Oscars','TMDb Score_y':'Score Golden Globes'})
nominated_median_score_all = []
for l in range(len(nominations_scores)):
    nominated_median_score_all.append(nominated_median_value)
nominations_scores['All films median score'] = nominated_median_score_all
```

```python
nominations_scores[['Score Oscars','Score Golden Globes','All films median score']].plot()
plt.xlabel('Years')
plt.ylabel('N. Films')
plt.title('Nominated films scores')
```
