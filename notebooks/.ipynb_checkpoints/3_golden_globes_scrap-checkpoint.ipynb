```python
#creamos el diccionario del que saldrá el DF con los nombres de las columnas y lo completamos con los datos de la web por cada año
```

```python
golden_globes = {
    'year_film': [],
    'year_award': [],
    'Golden Globes nominations':[],
    'nominee':[],
    'film':[],
    'Golden Globes winner':[]    
}
```

```python
years = [2021,2022,2023]
```

```python
for y in years:
    url = f'https://goldenglobes.com/winners-nominees/?y={y}'
    response = requests.get(url)
    sopa = BeautifulSoup(response.content, "html.parser")
    nom_cards = sopa.find_all('div',class_='c-nomination-card__content')
    year_award = y
    year_film = y -1
    for n in range(len(nom_cards)):
        category = nom_cards[n].find('h3', class_='u-type-h4 u-mb-2 u-mb-4@md').text.strip()
        winner_card = nom_cards[n].find_all('li', class_='c-nomination-card__nominee js-nomination-nominee is-active')[0].find_all('div')[1].find_all('a')
        nominee = winner_card[0].text.strip()
        film = ''
        if len(winner_card) > 1:
            film = winner_card[1].text.strip()
        winner = True
        golden_globes['Golden Globes nominations'].append(category)
        golden_globes['film'].append(film)
        golden_globes['nominee'].append(nominee)
        golden_globes['Golden Globes winner'].append(winner)
        golden_globes['year_award'].append(year_award)
        golden_globes['year_film'].append(year_film)
        nominees_f = nom_cards[n].find_all('li', class_='c-nomination-card__nominee js-nomination-nominee')
        for nf in range(len(nominees_f)):
            film = nominees_f[nf].find('div', class_='u-type-h11').text.strip()
            nominee = nominees_f[nf].find('a', class_='c-nomination-card__name c-nomination-card__link u-color-black-01').text.strip()
            winner = False
            golden_globes['Golden Globes nominations'].append(category)
            golden_globes['film'].append(film)
            golden_globes['nominee'].append(nominee)
            golden_globes['Golden Globes winner'].append(winner)
            golden_globes['year_award'].append(year_award)
            golden_globes['year_film'].append(year_film)
golden_globes
```

```python
df_golden_extra = pd.DataFrame(golden_globes)

df_golden_extra.tail(50)
```
